### Question 3). Graphical Representation to show Applications in terms of Location/County

```python
data[["Location", "CountyNumber"]]
fig, ax = plt.subplots(1, figsize=(100,30))
plt.suptitle(' Applications in terms of Location/County ', fontsize=30)
ax.bar(data['Location'].value_counts().index, 
       data['Location'].value_counts().values,
       ) 

plt.xlabel("Location/County", fontsize=30)
plt.ylabel("Number of Applications ", fontsize=30)
plt.show()
```

### 4. Graphical Representation to show Applications in terms of Level of education

### Question 5). Graphical representation to show how many students completed

```python
data["Completed "].unique() 
```

```python
my_labels = ["Completed", "Not Completed", "Completed"]
plt.suptitle(' Applications in terms of Location/County ', fontsize=30)
plt.pie(data.groupby("Completed ").size(),  labels = my_labels, autopct='%.0f%%')
plt.show()
```

 - From the above diagram, only 36% of the enrolled students completed the plogram
